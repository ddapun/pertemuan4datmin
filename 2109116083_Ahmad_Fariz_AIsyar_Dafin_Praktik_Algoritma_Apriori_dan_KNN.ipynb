```python
#membaca data
df = pd.read_csv("DatasetBaru - Sheet1.csv")
```

```python

# 1 Item set
df.sum()
```

```python
#melihat berapa banyak item yang ada dalam data
df.columns
```

```python
# 2 Item set
from itertools import permutations
# Extract unique items.

#membuat perbandingan data
rules = list(permutations(df.columns, 2))
rules
```

```python
#Filter yang ada isinya aja yang di hitung
rule_baru = []
for rule in rules:
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0:
    rule_baru.append(rule)
rule_baru
```

```python
#melihat berapa jumlah data berdasarkan perbandingan item yang tidak kosong

hitung = {} #Ganti dictionary
for rule in rule_baru:
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1)
      ])
  hitung[str(rule)] = banyak_data

hitung
```

```python
# Melihat berapa seringnya data muncul
# List support
supp = [i/len(df)*100 for i in hitung.values()]
print(supp)
print("---")
for item,support in hitung.items():
  print(f"{item} support = {support/len(df)*100}")
```

```python
# Melihat berapa bergantungnya sebuah data B ke data A
# List Confidens

print("---")
for item,ab in hitung.items():
  i = eval(item)
  print(f"{i[0]}->{i[1]} Confidence = {ab/df[i[0]].sum()*100}%")
```

```python
df.isna().sum()
```

```python
df2 = df.copy(deep=True)
df2.drop(columns=["Cabin"],inplace=True)
```

```python
df2.isna().sum()
```
